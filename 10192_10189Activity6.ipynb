```python
import seaborn as sns


# Convert timestamp to datetime
amazon_events_data['timestamp'] = pd.to_datetime(amazon_events_data['timestamp'])

# Fill or Drop Missing Values
amazon_events_data['shipping_speed'] = amazon_events_data['shipping_speed'].fillna('Unknown')  # Fill missing shipping speed
amazon_events_data['page_view_time'] = amazon_events_data['page_view_time'].fillna(0)  # Replace NaN with 0 for analysis

# Add Derived Columns
amazon_events_data['month'] = amazon_events_data['timestamp'].dt.month
amazon_events_data['year'] = amazon_events_data['timestamp'].dt.year

### **A. Popular Product Categories**
# Filter for purchases
purchases = amazon_events_data[amazon_events_data['event_type'] == 'purchase']

# Count purchases per product
popular_products = purchases.groupby('product_id').size().reset_index(name='purchase_count')

# Visualize Top 10 Products
top_products = popular_products.nlargest(10, 'purchase_count')
plt.figure(figsize=(10, 6))
sns.barplot(data=top_products, x='purchase_count', y='product_id', palette='viridis')
plt.title('Top 10 Products by Purchases')
plt.xlabel('Number of Purchases')
plt.ylabel('Product ID')
plt.show()

### **B. Cart Abandonment Patterns**
# Identify abandoned products
cart_abandoned = amazon_events_data[(amazon_events_data['event_type'] == 'cart') &
                                    (~amazon_events_data['product_id'].isin(purchases['product_id']))]

# Group by product
abandoned_summary = cart_abandoned.groupby('product_id').size().reset_index(name='abandonment_count')

# Visualize Top 10 Abandoned Products
# Debug abandoned cart data
print("Cart Abandoned DataFrame Preview:")
print(cart_abandoned.head())

print("\nAbandoned Summary Preview:")
print(abandoned_summary.head())

# Top 10 abandoned products
top_abandoned = abandoned_summary.nlargest(10, 'abandonment_count')

# Debug top_abandoned
print("\nTop Abandoned Products:")
print(top_abandoned)

# Check if top_abandoned is empty
if top_abandoned.empty:
    print("No abandoned products to display.")
else:
    # Plot the data
    plt.figure(figsize=(10, 6))
    sns.barplot(data=top_abandoned, x='abandonment_count', y='product_id', palette='mako')
    plt.title('Top 10 Abandoned Products')
    plt.xlabel('Abandonment Count')
    plt.ylabel('Product ID')
    plt.show()


### **C. Seasonal Trends**
# Group by month and event type
monthly_trends = purchases.groupby('month').size()

# Visualize Monthly Trends
plt.figure(figsize=(10, 6))
monthly_trends.plot(kind='bar', color='skyblue')
plt.title('Seasonal Purchase Trends')
plt.xlabel('Month')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=0)
plt.show()

### **D. Device Preferences**
# Group by device type
device_usage = amazon_events_data['device_type'].value_counts()

# Visualize Device Preferences
plt.figure(figsize=(8, 6))
device_usage.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Device Preferences')
plt.ylabel('')
plt.show()

### **E. Shipping Preferences**
# Group by shipping speed
shipping_speed = amazon_events_data['shipping_speed'].value_counts()

# Visualize Shipping Preferences
plt.figure(figsize=(10, 6))
sns.barplot(x=shipping_speed.values, y=shipping_speed.index, palette='coolwarm')
plt.title('Shipping Speed Preferences')
plt.xlabel('Count')
plt.ylabel('Shipping Speed')
plt.show()

### **F. Engagement Metrics**
# Analyze page view time
engagement = amazon_events_data.groupby('product_id')['page_view_time'].mean().reset_index(name='avg_page_view_time')

# Visualize Top 10 Engaging Products
top_engagement = engagement.nlargest(10, 'avg_page_view_time')
plt.figure(figsize=(10, 6))
sns.barplot(data=top_engagement, x='avg_page_view_time', y='product_id', palette='flare')
plt.title('Top 10 Engaging Products by Page View Time')
plt.xlabel('Average Page View Time')
plt.ylabel('Product ID')
plt.show()
```

```python
#Recommendation Logic:
def recommend_based_on_browsing(user_id):
    preferred_categories = users_df.loc[users_df['user_id'] == user_id, 'preferred_categories'].values[0]
    preferred_categories = eval(preferred_categories)  # Convert string to list
    recommendations = products_df[products_df['category'].isin(preferred_categories)].sample(5)
    return recommendations
```

```python
def recommend_based_on_abandoned_cart(user_id):
    # Get abandoned products for the specific user
    abandoned_products = abandoned_cart_data[abandoned_cart_data['user_id'] == user_id]['product_id'].unique()

    # Debugging: Check if the user has any abandoned products
    print(f"Abandoned products for user {user_id}: {abandoned_products}")

    if len(abandoned_products) == 0:
        return "No abandoned cart items found for this user."

    recommendations = products_df[products_df['product_id'].isin(abandoned_products)].sample(5)
    return recommendations
```

```python
user_features = users_df[['average_annual_spend', 'has_alexa_device', 'has_kindle']]

```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
users_df['cluster'] = kmeans.fit_predict(user_features)
```

```python

import matplotlib.pyplot as plt

# Calculate inertia for a range of cluster numbers
inertia = []
k_range = range(1, 11)  # Try different cluster numbers
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(user_features)
    inertia.append(kmeans.inertia_)

# Plot the inertia for each k
plt.plot(k_range, inertia, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()

print(f"Inertia: {kmeans.inertia_}")
```
